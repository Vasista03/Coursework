```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#1  Plot a line graph to display growth in population in the past 7 decades. 

# Create the dataframe
data = {
    'CensusYear': [1951, 1961, 1971, 1981, 1991, 2001, 2011],
    'Population': [361088000, 439235000, 548160000, 683329000, 846387888, 1028737436, 1210726932]
}

df = pd.DataFrame(data)

# Ensure Population is numeric (already numeric here)
df['Population'] = pd.to_numeric(df['Population'])

plt.plot(df['CensusYear'], df['Population'], marker='*')
plt.title("Population Growth by Census Year")
plt.xlabel("Year")
plt.ylabel("Population")
plt.show()
```

```python
#2  . Plot a line graph to show Sin Curve. (HOTS) Hint: Numpy has a function, numpy.sin() to find the sin values

sin_values = np.sin(df['CensusYear'] * np.pi / 180)  # converting years to radians

# Plot
plt.plot(df['CensusYear'], sin_values, marker='o')
plt.title("Sine Curve over Census Years")
plt.xlabel("Year")
plt.ylabel("sin(CensusYear in radians)")
plt.show()
```

```python
#3 Plot a line graph to show Cos Curve. (HOTS) Hint: Numpy has a function, numpy.cos() to find the cos values

cos_values = np.cos(df['CensusYear'] * np.pi / 180)  # converting years to radians

# Plot
plt.plot(df['CensusYear'], cos_values, marker='o')
plt.title("Cosine Curve over Census Years")
plt.xlabel("Year")
plt.ylabel("cos(CensusYear in radians)")
plt.show()
```

```python
#4  Plot a Bar Graph to show the number of boys in each class 6- 12. Data should be imagined by student

np.random.seed(42)
classes = ['6', '7', '8', '9', '10', '11', '12']
boys = np.random.randint(28,41,7)  # imagined numbers

# Bar plot
plt.bar(classes, boys, color='blue')
plt.title("Number of Boys in Each Class")
plt.xlabel("Class")
plt.ylabel("Number of Boys")
plt.show()
```

```python
#5  Plot a Bar Graph for Marks scored in different subjects. Data should be imagined.

np.random.seed(42)
subjects = ['Math', 'Science', 'English', 'History', 'Geography']
marks = np.random.randint(65,101,5)  # imagined marks

# Bar plot
plt.bar(subjects, marks, color='green')
plt.title("Marks Scored in Different Subjects")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.show()
```

```python
#6  Plot a histogram to show the number of employees arriving at the office between 7AM and 12Noon, using 1-hour intervals as bins

np.random.seed(42)
arrival_times = np.random.randint(7, 12, 50)  # 50 employees

# Histogram with 1-hour bins
plt.hist(arrival_times, bins=[7,8,9,10,11,12], color='orange', edgecolor='black')
plt.title("Employee Arrival Times (7 AM - 12 Noon)")
plt.xlabel("Hour of Arrival")
plt.ylabel("Number of Employees")
plt.show()
```
